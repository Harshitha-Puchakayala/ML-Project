```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
```

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
scaled_x = scaler.fit_transform(df.drop('Potability', axis=1))
```

```python
# Split the data into training and testing sets
x_tr, x_te, y_tr, y_te = train_test_split(scaled_x, df['Potability'], random_state=42, test_size=0.3)
```

```python
# k-Nearest Neighbors (KNN)
knn = KNeighborsClassifier()
knn_param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
gridsearch_knn = GridSearchCV(estimator=knn, param_grid=knn_param_grid, n_jobs=-1, verbose=4, cv=3)
gridsearch_knn.fit(x_tr, y_tr)
best_knn = gridsearch_knn.best_estimator_
predicted_y_knn = best_knn.predict(x_te)
```

```python
# Support Vector Machine (SVM)
svm = SVC()
svm_param_grid = {
    'C': [1.2, 1.5, 2.2, 3.5, 3.2, 4.1],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [1, 2, 4, 8, 10],
    'gamma': ['scale', 'auto']
}
```

```python
gridsearch_svm = GridSearchCV(estimator=svm, param_grid=svm_param_grid, n_jobs=-1, verbose=4, cv=3)
gridsearch_svm.fit(x_tr, y_tr)
best_svm = gridsearch_svm.best_estimator_
predicted_y_svm = best_svm.predict(x_te)
```

```python
# Logistic Regression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_tr, y_tr)
predicted_y_logistic = logistic_regression.predict(x_te)
```

```python
# Random Forest
random_forest = RandomForestClassifier()
random_forest_param_grid = {
    'n_estimators': [50, 100, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10]
}
```

```python
gridsearch_rf = GridSearchCV(estimator=random_forest, param_grid=random_forest_param_grid, n_jobs=-1, verbose=4, cv=3)
gridsearch_rf.fit(x_tr, y_tr)
best_rf = gridsearch_rf.best_estimator_
predicted_y_rf = best_rf.predict(x_te)
```

```python
# Gradient Boosting
gradient_boosting = GradientBoostingClassifier()
gradient_boosting_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
```

```python
gridsearch_gb = GridSearchCV(estimator=gradient_boosting, param_grid=gradient_boosting_param_grid, n_jobs=-1, verbose=4, cv=3)
gridsearch_gb.fit(x_tr, y_tr)
best_gb = gridsearch_gb.best_estimator_
predicted_y_gb = best_gb.predict(x_te)
```

```python
# XGBoost
xgb = XGBClassifier()
xgb.fit(x_tr, y_tr)
predicted_y_xgb = xgb.predict(x_te)
```

```python
# ID3 (Decision Tree)
id3 = DecisionTreeClassifier(criterion='entropy')
id3.fit(x_tr, y_tr)
predicted_y_id3 = id3.predict(x_te)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

classifiers = ['KNN', 'SVM', 'Logistic Regression', 'Random Forest', 'XGBoost', 'Gradient Boosting', 'ID3']
predictions = [predicted_y_knn, predicted_y_svm, predicted_y_logistic, predicted_y_rf, predicted_y_xgb, predicted_y_gb, predicted_y_id3]

for classifier, prediction in zip(classifiers, predictions):
    print(f"\n{classifier} Confusion Matrix:\n{confusion_matrix(y_te, prediction)}")

    acc = accuracy_score(y_te, prediction)
    print(f"{classifier} Accuracy: {acc}")

    precision = precision_score(y_te, prediction)
    print(f"{classifier} Precision: {precision}")

    recall = recall_score(y_te, prediction)
    print(f"{classifier} Recall: {recall}")

    f1 = f1_score(y_te, prediction)
    print(f"{classifier} F1 Score: {f1}")
```

```python
# Calculate the number of potable and non-potable instances in the dataset
potable_count = sum(y_te)  # Assuming y_te contains the true labels where 1 represents potable and 0 represents non-potable
not_potable_count = len(y_te) - potable_count

# Calculate the percentages
potable_percent = (potable_count / len(y_te)) * 100
not_potable_percent = (not_potable_count / len(y_te)) * 100

print(f"Percentage of potable instances: {potable_percent:.2f}%")
print(f"Percentage of non-potable instances: {not_potable_percent:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Define a color palette with different colors for each classifier
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral', 'lightblue', 'yellow']

# List to store accuracies
accuracies = []

# Iterate over classifiers and predictions
for classifier, prediction, color in zip(classifiers, predictions, colors):
    # Calculate accuracy
    acc = accuracy_score(y_te, prediction)
    accuracies.append(acc)

    # Plot accuracy with respective color
    plt.barh(classifier, acc, color=color)

# Plotting the accuracy comparison
plt.xlabel('Accuracy')
plt.ylabel('Classifier')
plt.title('Accuracy Comparison among Classifiers')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest accuracy at the top
plt.show()
```
